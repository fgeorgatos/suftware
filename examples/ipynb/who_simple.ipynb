```python
# Standard modules
import numpy as np
import matplotlib.pyplot as plt
import sys

# Import suftware 
sys.path.append('../../')
import suftware as su

### Load example data ###
data = su.example_data_1d('who.alcohol_consumption')

### Perform density estimation using DEFT 1D ###
Q = su.Deft1D(data)  

### Plot results ###
fig, ax = plt.subplots(figsize=[5,5])

# data histogram
ax.bar(Q.grid, Q.histogram, width=Q.grid_spacing, color='orange')

# posterior samples
ax.plot(Q.grid, Q.sample_values, color='dodgerblue', linewidth=1, alpha = .2)

# best fit density
ax.plot(Q.grid, Q.values, color='blue', linewidth=2)

# style plot
ax.set_xlim(Q.bounding_box)
ax.set_title('Alcohol consumption')
ax.set_xlabel('liters per year')
ax.set_yticks([])
plt.tight_layout()

# Save plot
fig.savefig('alcohol.png')
plt.show()
```
