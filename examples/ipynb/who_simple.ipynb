```python
# Standard modules
import numpy as np
import matplotlib.pyplot as plt
import sys

# Import suftware 
sys.path.append('../../')
import suftware as sw

dataset_name = 'who.alcohol_consumption'

### Load example data ###
data, details = sw.example_density_data(dataset_name)

### Perform density estimation using DEFT 1D ###
Q = sw.Density(data)  

### Plot results ###
fig, ax = plt.subplots(figsize=[5,5])

# data histogram
ax.bar(Q.grid, Q.histogram, width=Q.grid_spacing, color='orange')

# posterior samples
ax.plot(Q.grid, Q.sample_values, color='dodgerblue', linewidth=1, alpha = .2)

# best fit density
ax.plot(Q.grid, Q.values, color='blue', linewidth=2)

# style plot
ax.set_xlim(Q.bounding_box)
ax.set_title(details['description'])
ax.set_xlabel(details['units'])
ax.set_yticks([])
ax.tick_params('x', rotation=45)
plt.tight_layout()

# Save plot
plt.tight_layout()
plt.show()
```
