```python
# Standard modules
import numpy as np
import sys

# Import suftware 
sys.path.append('../../')
import suftware as sw
sw.enable_graphics()
%matplotlib inline
```

```python
# Retrieve data included with SUFTware
dataset = sw.ExampleDataset('who.alcohol_consumption')

# Perform one-dimensional density estimation
density = sw.DensityEstimator(dataset.data)

# Plot results and annotate with metadata
density.plot(title=dataset.description,
             xlabel=dataset.units,
             save_as=dataset.name + '.png')
```

```python
### On custom synthetic data ###

# Generate datas
data = np.random.gamma(5,1,size=100)

# Perform density estimation
density = sw.DensityEstimator(data)  

# Plot results
density.plot(title='Gaussian')
```
